# Visual Story Telling Part 2: Capital Metro Data

<div class="gray">
The file capmetro_UT.csv contains data from Austin's own Capital Metro bus network, including shuttles to, from, and around the UT campus. These data track ridership on buses in the UT area. Ridership is measured by an optical scanner that counts how many people embark and alight the bus at each stop. Each row in the data set corresponds to a 15-minute period between the hours of 6 AM and 10 PM, each and every day, from September through November 2018. The variables are:

- *timestamp*: the beginning of the 15-minute window for that row of data
- *boarding*: how many people got on board any Capital Metro bus on the UT campus in the specific 15 minute window
- *alightin*g*: how many people got off ("alit") any Capital Metro bus on the UT campus in the specific 15 minute window
- *day_of_week* day of the week (Monday, Tuesday, etc)
- *weekend*: indicator for whether a day is a weekend.
- *temperature*: temperature at that time in degrees F
- *hour_of_day*: on 24-hour time, so 6 for 6 AM, 13 for 1 PM, 14 for 2 PM, etc.
- *month*: July through December

Your task is to create a figure, or set of related figures, that tell an interesting story about Capital Metro ridership patterns around the UT-Austin campus during the semester in question. Provide a clear annotation/caption for each figure, but the figure(s) should be more or less stand-alone, in that you shouldn't need many, many paragraphs to convey its meaning. Rather, the figure together with a concise caption should speak for itself as far as possible.

You have broad freedom to look at any variables you'd like here -- try to find that sweet spot where you're showing genuinely interesting relationships among more than just two variables, but where the resulting figure or set of figures doesn't become overwhelming/confusing. (Faceting/panel plots might be especially useful here.)
</div>

```python
# Imporing Libraries and Reading in CSV Files

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from time import mktime
from datetime import datetime

df = pd.read_csv("data/capmetro_UT.csv")
df.head()

# Check if any of the data types don't match what I expect 
df.dtypes

# Check missing values
df.isna().sum()

# The length of the dataset is:
len(df)
```

**Dataset Preparation**:
After reviewing the dataset, "timestamp" needs to be converted from an "object" to a datetime variable in order to work with it more effectively. Some of the information in "timestamp" is already encoded in "hour_of_day", "month", "day_of_week" and "weekend", but we'll convert it just to be safe.

There are no missing values and the dataset contains 5824 entries. 

```python
def parse_date(date_str):
    entry = date_str[0:20]
    return datetime.strptime(entry, "%Y-%m-%d %H:%M:%S")

dates = [parse_date(entry) for entry in df["timestamp"]]

df["Time"] = dates

df = df.drop(columns = "timestamp")

df['hour'] = df['Time'].dt.hour
df['day_of_week'] = df['Time'].dt.day_name()
```

We are provided with two metrics, boarding and alighting. Let's explore how these change as the day progresses. We can split this information by month to see if there are any differences as the semester progresses. 

Then we can look at any changes between days.

```python
hour_of_day = df.groupby(["month", "hour_of_day"])["boarding"].mean().reset_index()
sns.scatterplot(x = "hour_of_day", y = "boarding", data = hour_of_day, hue = "month", style = "month")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Boarders")
plt.title("Average Number of Boarders by Hour")
```

```python
hour_of_day = df.groupby(["month", "hour_of_day"])["alighting"].mean().reset_index()
sns.scatterplot(x = "hour_of_day", y = "alighting", data = hour_of_day, hue = "month", style = "month")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Alighters")
plt.title("Average Number of Alighters by Hour")
```

```python
al_heatmap_data = df.pivot_table(index='day_of_week', columns='hour', values='alighting', aggfunc='sum')
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
al_heatmap_data = al_heatmap_data.reindex(days_order)

plt.figure(figsize=(14, 8))
sns.heatmap(al_heatmap_data, cmap='coolwarm', cbar_kws={'label': 'Number of Alightings'})
plt.title('Alightings by Day of the Week and Hour')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
bo_heatmap_data = df.pivot_table(index='day_of_week', columns='hour', values='boarding', aggfunc='sum')
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
bo_heatmap_data = bo_heatmap_data.reindex(days_order)

plt.figure(figsize=(14, 8))
sns.heatmap(bo_heatmap_data, cmap='coolwarm', cbar_kws={'label': 'Number of Boardings'})
plt.title('Boardings by Day of the Week and Hour')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
plt.show()
```

Question: why are boardings and alightings only peaking at one time of the day? Shouldn't the people who alight in the morning have boarded some time in the minutes/ hour before? The same goes for the peak in boardings in the afternoon; when are these people getting off the bus?

We will look at changes in passenger count on the bus to get a single metric that can be used to compare "busy" or "not busy" times of day.

```python
df["Bus_Count"] = 0
for date in df["Time"].unique():
    bus_count = 0
    data = df[df["Time"] == date]
    for i in range(len(data)):
        bus_count += data.iloc[i]["boarding"]
        bus_count -= data.iloc[i]["alighting"]
        df.loc[data.index[i], "Bus_Count"] = bus_count

df.head()

# Note: this all assumes that the bus is empty when the day starts. A quick review of "Passenger_Count" reveals that this is not the case.
# There are times when someone gets off a bus that should've been empty, or the bus still has passengers on it at the end of the day.
```

Bus count doesn't seem to be a good measure of people getting on or off the bus since we can't assume the bus was empty when each day starts.

Instead, let's create another variable called passenger difference that only looks at the number of people boarding the bus minus those alighting.

We'll plot some more scatterplots and heatmaps of how month, day of the week and time of day change passenger difference. We'll also look at one more scatterplot to show how day of the week and weekend affect passenger difference.


```python
df["passenger_diff"] = df["boarding"] - df["alighting"]
passenger_rate = df.groupby(["month", "hour_of_day"])["passenger_diff"].mean()
passenger_rate = passenger_rate.reset_index()

sns.scatterplot(x = "hour_of_day", y = "passenger_diff", data = passenger_rate, hue = "month", style = "month")
plt.xlabel("Hour of Day")
plt.ylabel("Passenger Difference")
plt.title("Passenger Difference by Time of Day")
```

```python
board_by_weekday = df.groupby(["weekend", "hour_of_day"])["passenger_diff"].mean()
board_by_weekday = board_by_weekday.reset_index()

sns.scatterplot(x = "hour_of_day", y = "passenger_diff", data = board_by_weekday, hue = "weekend")

# Time of day doesn't seem to affect traveller choices on the weekend as much as it does during the week. This makes sense; people use the bus to get
# to campus for work/ school during the week but may use it more sporadically during the weekend.
```

```python
pd_heatmap_data = df.pivot_table(index='day_of_week', columns='hour', values='passenger_diff', aggfunc='sum')
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pd_heatmap_data = pd_heatmap_data.reindex(days_order)

plt.figure(figsize=(14, 8))
sns.heatmap(pd_heatmap_data, cmap='coolwarm', cbar_kws={'label': 'Passenger Difference'})
plt.title('Passenger Difference by Day of the Week and Hour of the Day')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
plt.show()
```

Passenger difference is negative in the mornings; this matches with our previous findings that alightings were high in the morning and boardings were high in the afternoon. We wanted passenger difference to be a measure of people getting on/ off the bus. We assume the people who got on the bus at 8am are leaving the bus some time in the next hour, but for some reason the optical scanner isn't picking them up twice. 

Taking the absolute value of the difference between boarding and alighting customers will give us one metric that we can use to determine when the bus is busy or empty. We will name this new variable *"bus_activity"*.

```python
df["bus_activity"] = abs(df["passenger_diff"])

bus_activity = df.groupby(["month", "hour_of_day"])["bus_activity"].mean()
bus_activity = bus_activity.reset_index()

sns.scatterplot(x = "hour_of_day", y = "bus_activity", data = bus_activity, hue = "month", style = "month")
plt.xlabel("Hour of Day")
plt.ylabel("Bus Activity")
plt.title("Bus Activity by Time of Day")
```

```python
heatmap_data = df.pivot_table(index='day_of_week', columns='hour', values='bus_activity', aggfunc='sum')
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(days_order)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', cbar_kws={'label': 'Bus Activity'})
plt.title('Bus Activity by Day of the Week and Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

Bus_activity has given us a final measure of how busy the bus is. There are two peaks, one in the morning and one in the afternoon. This aligns with the times of day when people go to work, attend class etc.

Time of day clearly has a huge influence on when people use the bus. Temperature is also a component of our dataset, and we wanted to explore how this affects when people use the bus. Specifically, we wanted to see whether unexpectedly warm or cold days influence bus usage. 

```python
# temp_averages calculates the average temperature for each hour of the day by month.
temp_averages = df.groupby(["month", "hour_of_day"])["temperature"].mean()
df2 = df.merge(temp_averages, on = ["month", "hour_of_day"]).rename(columns = {"temperature_y":"mean"})
df2["temp_diff"] = df2["temperature_x"] - df2["mean"]

print("The number of days that are colder than average are ", len(df2[df2["temp_diff"] < 0]))
print("The number of days that are hotter than average are ", len(df2[df2["temp_diff"] > 0]))

# We want to look at exclusively large outliers, anything warmer than one standard deviation above mean or colder than one standard deviation below mean.

print("The standard deviation for colder days is:", df2[df2["temp_diff"] < 0]["temp_diff"].std())
print("The standard deviation for colder days is:", df2[df2["temp_diff"] > 0]["temp_diff"].std())
```

```python
super_cold = df2[df2["temp_diff"] < -5.86]

super_cold = super_cold.groupby(["hour_of_day", "month"])["bus_activity"].mean()
super_cold = super_cold.reset_index()


# Plot the scatter plot for super cold days
sns.scatterplot(x="hour_of_day", y="bus_activity", data=super_cold, hue="month", style="month")

overall_mean = df2.groupby("hour_of_day")["bus_activity"].mean().reset_index()
sns.lineplot(x="hour_of_day", y="bus_activity", data=overall_mean, color="red", linestyle="--", label="Overall Mean")

# Add labels and title
plt.xlabel("Hour of Day")
plt.ylabel("Bus Activity")
plt.title("Bus Activity on Unexpectedly Cold Days")
plt.legend()

# Show plot
plt.show()
```

```python
super_hot = df2[df2["temp_diff"] > 4.69]


super_hot = super_hot.groupby(["hour_of_day", "month"])["bus_activity"].mean()
super_hot = super_hot.reset_index()

sns.scatterplot(x = "hour_of_day", y = "bus_activity", data = super_hot, hue = "month")

plt.xlabel("Hour of Day")
plt.ylabel("Bus Activity")
plt.title("Bus Activity on Unexpectedly Warm Days")

overall_mean = df2.groupby("hour_of_day")["bus_activity"].mean().reset_index()
sns.lineplot(x="hour_of_day", y="bus_activity", data=overall_mean, color="red", linestyle="--", label="Overall Mean")
```
